# Extraction des contours 

```python
class ExtractEdges:
    def __init__(self,):
        pass
    def filtre(img, kernel):
        newimg = np.zeros(img.shape, dtype=np.float32)

        kernel_height, kernel_width = kernel.shape
        kh, kw = kernel_height // 2, kernel_width // 2  

        for i in range(kh, img.shape[0] - kh):
            for j in range(kw, img.shape[1] - kw):
                sum_value = 0
                for ki in range(kernel_height):
                    for kj in range(kernel_width):
                        sum_value += kernel[ki, kj] * img[i + ki - kh, j + kj - kw]

                newimg[i, j] = sum_value
        return newimg
    def sobel(img):
        # gradient selon x 
        kernel = np.array([[-1, 0, 1], [-2,0,2], [-1, 0, 1]])/5
        gradx = filtre(img, kernel)
        # gradient selon y 
        kernel = np.array([[1, 2, 1], [0,0,0], [-1, -2, -1]])/5
        grady = filtre(img, kernel)    
        # image final 
        newimg = np.sqrt(gradx**2 + grady**2)
        newImg = cv2.threshold(newimg, 128, 1,cv2.THRESH_BINARY )
        return newImg[1]
    def prewitt(img):
        # gradient selon x 
        kernel = np.array([[-1, 0, 1], [-1,0,1], [-1, 0, 1]])/3
        gradx = filtre(img, kernel)


        # gradient selon y 
        kernel = np.array([[-1, -1, -1], [0,0,0], [1, 1, 1]])/3
        grady = filtre(img, kernel)

        # image final 
        newimg = np.sqrt(gradx**2 + grady**2)
        newImg = cv2.threshold(newimg, 90, 1,cv2.THRESH_BINARY)
        return newImg[1]
    def robinson(img, level):
        if level == 3:
            c = 1
            n = 3
        elif level == 5:
            c = 2
            n = 5
        else:
            return "the level should be 3 or 5"
        # gradient niveau 1 
        kernel = np.array([[1,c,1], [0, 0, 0], [-1, -c, -1]])/n
        grad1 = filtre(img, kernel)
        # gradient niveau 2 
        kernel = np.array([[c,1,0], [1, 0, -1], [0, -1, -c]])/n
        grad2 = filtre(img, kernel)
        # gradient niveau 3 
        kernel = np.array([[1,0,-1], [c, 0, -c], [1, 0, -1]])/n
        grad3 = filtre(img, kernel)
        # gradient niveau 4 
        kernel = np.array([[-0,-1,-c], [1, 0, -1], [c, 1, 0]])/n
        grad4 = filtre(img, kernel)
        # gradient niveau 5 
        kernel = np.array([[-1,-c,-1], [0, 0, 0], [1, c,1]])/n
        grad5 = filtre(img, kernel)
        # gradient niveau 6 
        kernel = np.array([[-c,-1,0], [-1, 0, 1], [0, 1, c]])/n
        grad6 = filtre(img, kernel)
        # gradient niveau 7
        kernel = np.array([[-1,0,1], [-c, 0, c], [-1, 0, 1]])/n
        grad7 = filtre(img, kernel)
        # gradient niveau 8 
        kernel = np.array([[0,1,c], [-1, 0, 1], [-c, -1, 0]])/n
        grad8 = filtre(img, kernel)

        # final image
        newimg = np.sqrt(grad1**2 + grad2**2 + grad3**2 + grad4**2 + grad5**2 + grad6**2 + grad7**2 + grad8**2)
        _, newImg = cv2.threshold(newimg, 90, 1,cv2.THRESH_BINARY)
        return newImg
    def roberts(img):
        # gradient selon x
        kernel = np.array([[1, 0], 
                         [0, -1]])
        gradx = filtre(img, kernel)
        # gradient selon y
        kernel = np.array([[0, 1], 
                         [-1, 0]])
        grady = filtre(img, kernel)
        # image final 
        newimg = np.sqrt(gradx**2 + grady**2)
        _ , newImg = cv2.threshold(newimg, 90, 1,cv2.THRESH_BINARY)
        return newImg
    def laplacien(img, kernelType):
        # 
        if kernelType == 1:
            kernel = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])
        elif kernelType == 2:
            kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
        elif kernelType == 3:
            kernel = np.array([[1, -2, 1], [-2, 4, -2], [1, -2, 1]])
        else:
            return "the kernel type should be 1,2 or 3"
        img = filtre(img, kernel)
        _ , newimg = cv2.threshold(img, 90, 1,cv2.THRESH_BINARY)

        return newimg 
```

# Morphologie

```python
class Morphologie:
    def __init__(self):
        pass
    def Erosion(img, kernel, iteration):
        
        imgBinaire = cv2.threshold(img, 128, 255,cv2.THRESH_BINARY)

        imgBinaire = imgBinaire[1]

        newImg = imgBinaire.copy()


        kw, kh = kernel.shape[0]//2, kernel.shape[1]//2

        k = kernel[kw, kh]

        for k in range(iteration):

            for i in range(kw, imgBinaire.shape[1]-kw):

                for j in range(kh, imgBinaire.shape[0]-kh):

                    if imgBinaire[i,j] == k:

                        if not np.array_equal(kernel, imgBinaire[i-1:i+2,j-1:j+2]):
                            newImg[i,j] = 255


        return newImg
    def Dilatation(img, kernel, iteration):
    
        imgBinaire = cv2.threshold(img, 128, 255,cv2.THRESH_BINARY)

        imgBinaire = imgBinaire[1]

        newImg = imgBinaire.copy()


        kw, kh = kernel.shape[0]//2, kernel.shape[1]//2

        k = kernel[kw, kh]

        for k in range(iteration):

            for i in range(kw, imgBinaire.shape[1]-kw):

                for j in range(kh, imgBinaire.shape[0]-kh):

                    if imgBinaire[i,j] == k:

                        if not imgBinaire[i-1:i+2,j-1:j+2].all():
                            newImg[i,j] = 0
        return newImg
    def Ouverture(img,kernel, intertion):
        # application de l'erosion
        imgErode = Erosion(img,kernel, intertion)
        # application de la dilatation 
        imgDilate = Dilatation(imgErode,kernel, intertion)
        return imgDilate
    def fermeture(img,kernel, iteration):
        # application de la dilatation 
        imgDilate = Dilatation(img,kernel, iteration)
        # application de l'erosion
        imgErode = Erosion(imgDilate,kernel, iteration)

        return imgErode
```

# Filtrage

```python
class Filtrage:
    def __init__(self):
        pass
    def Lissage(img, kernel_size = 3):
        
        kernel = np.ones((kernel_size,kernel_size))
        
        imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        newimg = np.zeros(imgGray.shape, dtype=np.float32)

        kernel_height, kernel_width = kernel.shape
        kh, kw = kernel_height // 2, kernel_width // 2  

        for i in range(kh, imgGray.shape[0] - kh):
            for j in range(kw, imgGray.shape[1] - kw): 
                sum_value = 0
                for ki in range(kernel_height):
                    for kj in range(kernel_width):
                        sum_value += kernel[ki, kj] * imgGray[i + ki - kh, j + kj - kw]

                newimg[i, j] = sum_value/(kernel.shape[0]**2)

        newimg = np.clip(newimg, 0, 255).astype(np.uint8)

        return newimg
    def filremedian(img, size):
        imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        newimg = imgGray.copy()
        t = size // 2
        for i in range(t, imgGray.shape[0]-t):
            for j in range(t , imgGray.shape[1]-t):
                m = np.sort(imgGray[i-t:i+t+1, j-t:j+t+1].flatten())
                newimg[i,j] = m[(size**2)//2]

        return newimg
            
```

# Segmentation

```python
class Segmentation:
    def __init__(self):
        pass
    def croissanceReg(img, gx, gy):
        newimg = img.copy()
        is_processed = np.zeros((img.shape[0], img.shape[1]))
        stack = [[gx, gy]]
        is_processed[gx, gy] = 1
        newimg[gx, gy] = [255, 0, 0]
        while stack:
            gx, gy = stack.pop()  

            if gy > 0 and (img[gx, gy] == img[gx, gy - 1]).all() and is_processed[gx, gy - 1] != 1:
                stack.append([gx, gy - 1])
                newimg[gx, gy - 1] = [255, 0, 0]
                is_processed[gx, gy - 1] = 1
            if gx < img.shape[0] - 1 and (img[gx, gy] == img[gx + 1, gy]).all() and is_processed[gx + 1, gy] != 1:
                stack.append([gx + 1, gy])
                newimg[gx + 1, gy] = [255, 0, 0]
                is_processed[gx + 1, gy] = 1    
            if gx > 0 and (img[gx, gy] == img[gx - 1, gy]).all() and is_processed[gx - 1, gy] != 1:
                stack.append([gx - 1, gy])
                is_processed[gx - 1, gy] = 1
                newimg[gx - 1, gy] = [255, 0, 0]

            if gy < img.shape[1] - 1 and (img[gx, gy] == img[gx, gy + 1]).all() and is_processed[gx, gy + 1] != 1:
                stack.append([gx, gy + 1])
                newimg[gx, gy + 1] = [255, 0, 0]
                is_processed[gx, gy + 1] = 1



        return newimg
```
